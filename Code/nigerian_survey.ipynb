```python
import csv

responses = []

for i in range(1, 100):
    print(f"\nPerson {i}: ")

    name = input("Enter your name: ")
    gender = input("Enter your gender (Male/Female): ")
    age = int(input("Enter your age: "))
    address = input("Enter your address: ")
    status = input("Enter your marital status (Single/Married): ")
    occupation = input("Enter occupation: ")
    problems = input("Enter your problem: ")
    
    person_info = {
         "Name": name,
         "Gender": gender,
         "Age": age,
         "Address": address,
         "Status": status,
         "Occupation": occupation,
         "Problems": problems
    }

    responses.append(person_info)

    # Save to CSV file after each entry
    with open("nigerian_survey.csv", mode='w', newline='', encoding='utf-8') as file:
        fieldnames = ["Name", "Gender", "Age", "Address", "Status", "Occupation", "Problems"]
        writer = csv.DictWriter(file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(responses)

    print("\nSurvey Summary: ")
    print(f"Total responses collected: {len(responses)}")
    print("Data saved to 'nigerian_survey.csv'")
```

```python
import csv
# To read csv file
from collections import Counter
# to count fraquency of each problem
import matplotlib.pyplot as plt
# to create the bar chart
import numpy as np
## for numerical operations,especially bar positioning

# Lists to store problems based on gender
male_problems = []
female_problems = []

# Read the CSV file
with open("nigerian_survey.csv", mode='r', encoding='utf-8') as file:
    reader = csv.DictReader(file)
    for row in reader:
        problem = row["Problems"].strip().lower()
        gender = row["Gender"].strip().lower()
        
        if gender == "male":
            male_problems.append(problem)
        elif gender == "female":
            female_problems.append(problem)

# Count the frequency of problems for males and females
male_problem_counts = Counter(male_problems)
female_problem_counts = Counter(female_problems)

# Get the top 3 problems for males and females
top_3_male_problems = male_problem_counts.most_common(3)
top_3_female_problems = female_problem_counts.most_common(3)

# Separate the problem names and their counts for males and females
male_problems_titles = [item[0].title() for item in top_3_male_problems]
male_counts = [item[1] for item in top_3_male_problems]

female_problems_titles = [item[0].title() for item in top_3_female_problems]
female_counts = [item[1] for item in top_3_female_problems]

# Set up the positions for the bars
x = np.arange(len(male_problems_titles))  # the label locations
width = 0.35  # the width of the bars
## arangin the bar chart starting from 0

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 6))
# it means the figure will be 10 inches wide 6
## ax is used to control whats drawn on the chart(like bars, labe , and title)

# Plot the bars for males and females
rects1 = ax.bar(x - width/2, male_counts, width, label='Males', color='blue')
## it draws the bar for male  and the width will shift the male bar a little to the left to the female bar so it will not overlap
rects2 = ax.bar(x + width/2, female_counts, width, label='Females', color='orange')

# Add some text for labels, title, and custom x-axis tick labels
ax.set_xlabel('Problems')
ax.set_ylabel('Count')
ax.set_title('Top 3 Nigerian Problems by Gender')
ax.set_xticks(x)
ax.set_xticklabels(male_problems_titles)
ax.set_yticks(np.arange(0, 30, 10))
ax.legend()
## it shows a color label for male and female

# Add values on top of each bar for clarity
def add_values(rects, ax):
    for rect in rects:
        height = rect.get_height()
        ax.annotate("{}".format(height),
                    xy=(rect.get_x() + 
                    rect.get_width() /2, height),
                    xytext=(0,3),
                    textcoords="offset points",
                    ha= "center" , va="bottom")
add_values(rects1, ax)
add_values(rects2, ax)

# rotate x for better readability
plt.xticks(rotation=45, ha="right")

# Show the chart
plt.tight_layout()
# avoids label cut off
plt.show()
```
